## ¿Qué es SVM?
SVM (Support Vector Machine) es un algoritmo de aprendizaje supervisado que se utiliza para clasificación y regresión. Su objetivo principal es encontrar el hiperplano óptimo que separa las clases con el mayor margen posible.

### Separación lineal

Dado un conjunto de datos etiquetado, SVM encuentra el hiperplano que maximiza la distancia (margen) entre las clases.

### Vectores soporte

Son los puntos que están más cerca del hiperplano y determinan su posición.

### Margen

Es la distancia entre el hiperplano y los vectores soporte más cercanos de cada clase. SVM maximiza este margen.

### Clasificación no lineal

SVM puede usar kernels para transformar los datos en espacios de mayor dimensión donde sí puedan separarse linealmente.

# Componentes clave
| Elemento            | Descripción                                                           |
| ------------------- | --------------------------------------------------------------------- |
| **Hiperplano**      | Frontera de decisión                                                  |
| **Margen**          | Distancia a los vectores soporte                                      |
| **C**               | Parámetro de penalización (tradeoff entre margen y errores)           |
| **Kernel**          | Función que transforma los datos (lineal, RBF, polinomial, sigmoid)   |
| **Slack variables** | Permiten ciertos errores si los datos no son perfectamente separables |


# Fórmulas clave

### Hiperplano lineal:
f(x)=w⋅x+b

### Condición de clasificación correcta:
yi​(w⋅xi​+b)≥1



```python
from sklearn.datasets import make_blobs
from sklearn.svm import SVC
import matplotlib.pyplot as plt
import numpy as np

# Crear datos de ejemplo
X, y = make_blobs(n_samples=100, centers=2, random_state=6)

# Entrenar SVM
model = SVC(kernel='linear')
model.fit(X, y)

# Visualizar
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='bwr')
ax = plt.gca()

# Dibujar hiperplano
xlim = ax.get_xlim()
xx = np.linspace(*xlim)
w = model.coef_[0]
b = model.intercept_[0]
yy = -(w[0] * xx + b) / w[1]
plt.plot(xx, yy, 'k--')
plt.title("SVM con separación lineal")
plt.show()
```

# Aplicación

Esta aplicación utiliza el algoritmo de Support Vector Machine (SVM) para clasificar imágenes de dígitos manuscritos del 0 al 9, basándose en el conjunto de datos clásico Digits de scikit-learn.

```python
# Aplicacion Practica
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report

# Cargar dataset
digits = datasets.load_digits()
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.2, random_state=42)

# Entrenar modelo SVM con kernel RBF
model = SVC(kernel='rbf', C=10, gamma=0.001)
model.fit(X_train, y_train)

# Evaluar
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```
