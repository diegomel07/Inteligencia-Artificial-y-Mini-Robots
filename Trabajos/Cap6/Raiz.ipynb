Primero realizamos una prueba con un modelo sencillo, regresion lineal.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Paso 1: Generar dataset
X = np.linspace(0, 1000, 10000).reshape(-1, 1)
y = np.sqrt(X)

# Paso 2: Entrenar modelo lineal
model = LinearRegression()
model.fit(X, y.ravel())

# Paso 3: Probar con 10 ejemplos
X_test = np.array([4, 16, 25, 36, 49, 64, 81, 100, 121, 144]).reshape(-1, 1)
y_real = np.sqrt(X_test)
y_pred = model.predict(X_test)

# Paso 4: Crear tabla de resultados
resultados = [(int(x), float(real), float(pred)) for x, real, pred in zip(X_test.flatten(), y_real.flatten(), y_pred)]
df_resultados = pd.DataFrame(resultados, columns=["x", "Raíz real", "Raíz predicha (ML)"])
df_resultados.index = np.arange(1, len(df_resultados) + 1)
df_resultados.index.name = "Nº"
df_resultados = df_resultados.round(4)

# Paso 5: Mostrar tabla
print(df_resultados)

# Paso 6: Gráfico comparativo
plt.figure(figsize=(10, 5))
plt.plot(X_test, y_real, label="Raíz real", marker='o')
plt.plot(X_test, y_pred, label="Predicha (Regresión Lineal)", marker='x')
plt.title("Comparación de Raíz Cuadrada: Real vs. Regresión Lineal")
plt.xlabel("x")
plt.ylabel("sqrt(x)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

Finalmente observamos como nuestro modelo no se ajusta a nuestros datos generando predicciones erroneas

Por otro lado, probamos con un modelo mas robusto, en este caso es Random Forest.
Concluimos observando como nuestro modelo preduce casi con 100% de seguridad un resultado correcto.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor

# Paso 1: Generar dataset
X = np.linspace(0, 1000, 10000).reshape(-1, 1)
y = np.sqrt(X)

# Paso 2: Entrenar modelo
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X, y.ravel())  # ravel() convierte y a 1D

# Paso 3: Probar con 10 ejemplos
X_test = np.array([4, 16, 25, 36, 49, 64, 81, 100, 121, 144]).reshape(-1, 1)
y_real = np.sqrt(X_test)
y_pred = model.predict(X_test)

# Paso 4: Crear tabla de resultados
resultados = [(int(x), float(real), float(pred)) for x, real, pred in zip(X_test.flatten(), y_real.flatten(), y_pred)]
df_resultados = pd.DataFrame(resultados, columns=["x", "Raíz real", "Raíz predicha (ML)"])
df_resultados.index = np.arange(1, len(df_resultados) + 1)
df_resultados.index.name = "Nº"
df_resultados = df_resultados.round(4)

# Paso 5: Mostrar tabla
print(df_resultados)

# Paso 6: Gráfico comparativo
plt.figure(figsize=(10, 5))
plt.plot(X_test, y_real, label="Raíz real", marker='o')
plt.plot(X_test, y_pred, label="Raíz predicha (ML)", marker='x')
plt.title("Comparación de Raíz Cuadrada: Real vs. ML")
plt.xlabel("x")
plt.ylabel("sqrt(x)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
